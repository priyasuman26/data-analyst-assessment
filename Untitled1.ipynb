```python
customers = pd.read_csv("customers (1).csv")
subscriptions = pd.read_csv("subscriptions.csv")
events = pd.read_csv("events.csv")
```

```python
datasets = {
    "customers": customers,
    "subscriptions": subscriptions,
    "events": events
}
```

```python
for name, df in datasets.items():
    print(f"\n--- {name.upper()} ---")
    print("Shape:", df.shape)
    print("Columns:", df.columns.tolist())
    print("\nInfo:")
    print(df.info())
```

```python
for name, df in datasets.items():
    print(f"\n--- {name.upper()} QUALITY CHECKS ---")
    print("Missing values:")
    print(df.isnull().sum())
    print("\nDuplicate rows:", df.duplicated().sum())
```

```python
customers["customer_id"].is_unique
subscriptions["subscription_id"].is_unique
events["event_id"].is_unique
```

```python
subscriptions_fk_issues = subscriptions[
    ~subscriptions["customer_id"].isin(customers["customer_id"])
]

events_fk_issues = events[
    ~events["customer_id"].isin(customers["customer_id"])
]

len(subscriptions_fk_issues), len(events_fk_issues)
```
